## 불인덱싱

```python
# 온도(temp)가 30도 이상인 데이터 추출
#df[df['temp']>=30]
w=df['temp']>=30
df[w]
```

```python
#최대풍속(max_wind)이 5이하인 데이터 추출
df[df['max_wind']<=5]
```

```python
# 판다스 논리연산자 : & and | or ~ 파이썬 기초와는 다르다
# 온도가 30도 이상이고 최대풍성이 9이상인 데이터추출
# 논리 비교 연산자 있을땐 비교쪽에 괄호필수
w=(df['temp']>=30) & (df['max_wind']>=9)
df[w]
```

```python
# 가장 더웠던 날은 언제일까
w = df['temp'].max()==df['temp']
df[w]
```

```python
# 가장 추운날
w=df['temp'].min() == df['temp']
df[w]
```
