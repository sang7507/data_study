```python
#1. if 문 :
temp = 9
if temp<=10 :
  print("난방을 가동하세요")
```

```python
#2. if - else : 합 불
score = 70
if score>= 60:
  print("pass")
else :
  print("fail")
```

```python
#3. if~elif~elif ......else문 : 성적프로그램(A,B,C,F)
score = 90
if score >=90 :
  print("A학점")
elif score >= 80:
  print("B학점")
elif score >= 70:
  print("C학점")
else :
  print("F학점")
```

```python
# 효율적인 코드
score = int(input("점수를 입력하세요 :"))
if score>= 90:
  grade = 'A'
elif score>= 80:
  grade = 'B'
elif score>= 70:
  grade = 'C'
else :
  grade = 'F'
# 출력
print(f'{score}=>{grade}학점')
```

```python
# 도전문제 : elif문을 사용하여 이번 학기 이수과목(3과목 점수 입력)
# 평균이 90점 이상이면 전액 장학금,80점이상이면 반액장학금 나머지 미지급
#입력
score1 = int(input("첫과목 점수 :"))
score2 = int(input("두번쨰 과목 점수 :"))
score3 = int(input("세번째 과목 점수 :"))
#처리
sum = score1+score2+score3
avg = sum/3
if avg >= 90 :
  result = "전액장학금 대상"
elif avg >= 80 :
  result = "반액장학금 대상"
else :
  result = "대상이 아닙니다."

#출력
print(f"통보결과 : {result}")
```

```python
# 매개변수로 정수를 입력받아 +5결과를 반환하는 함수정의
def plus_five(x) :
  return x+5

plus_five(10)
```
